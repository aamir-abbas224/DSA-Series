```python
#Binary Search : Its goal is to find the position of a target value in a sorted array.It works by repeatedly dividing the search range in half until the target is found (or the range is empty).

#recursive binary search: Cleaner and elegant and uses stack frames and is slightly slower due to recursive overhead O(n)
def binary_search(data,target,low,high):
    
    if low > high:
        return False
    else:
        mid = (low+high) // 2
        if target == data[mid]:
            return True
        elif target < data[mid]:
            return binary_search(data,target,low,mid-1)
        else:
            return binary_search(data, target, mid + 1, high)
```

```python
#iterative binary search: slightly faster and is preferred in real-world applications. No risk of Stack overflow and the memory usage is O(1)
def binary_search(data,target):
    low = 0 #must always be 0 to start from first index
    high = len(data)-1 #the last index
    
    #the main logic to start with while loop
    while low<=high:
        mid = len(data) // 2 #the middle element of the dataset
        
        if target == data[mid]:
            return True
        else:
            if target > data[mid]:
                high = mid-1 
            else:
                low = mid+1
                
    return False        

arr = [1, 3, 5, 7, 9]
sorted(arr) #requirement for example
print(binary_search(arr, 5))
```

```python
def linear_search(arr, target):
    for i in arr:
        if i == target:
            return True
    return False

arr = [7, 1, 9, 3, 5]
print(linear_search(arr, 5))
```
